# Energy consumptions in Portugal in the last three decades and the impact of the Covid-19 pandemic in them

## Introduction

With the development of the Pandemic caused by Covid-19, a lot of people started working from home due to lockdown in Portugal. In this small study, our main goal is to evaluate the energy consumptions in the last 27 years, in order to determine how remote work has changed the energy consumptions of families, industries and governmental sectors.

After this evaluation, we will also do a forecast for the next 5 years, in order to show a possible prediction. This prediction will take into account the country slowly opening and coming out of Lockdown, as we cannot predict the existance of some highly lethal mutation of the virus that will cause another lockdown of the country.

## Loading data and imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('energy.csv')
df.set_index('Year', inplace=True)
df
```

## Data Analysis

```python
df.describe()
```

```python
fig, ax =plt.subplots(4,2, figsize=(20, 20))
sns.lineplot(data=df, x='Year', y='Total', ax=ax[0][0])
sns.lineplot(data=df, x='Year', y='Domestic', ax=ax[0][1])
sns.lineplot(data=df, x='Year', y='Non-domestic', ax=ax[1][0])
sns.lineplot(data=df, x='Year', y='Industry', ax=ax[1][1])
sns.lineplot(data=df, x='Year', y='Agriculture', ax=ax[2][0])
sns.lineplot(data=df, x='Year', y='StreetLights', ax=ax[2][1])
sns.lineplot(data=df, x='Year', y='GovernmentBuildings', ax=ax[3][0])
```

For a better understanding, the Energy units in this graph is in kWh per capita for the whole year. Analyzing the graphs allows us to see that the total energy consumption is tendentially increasing to these days since the early 1990s. While it has lowered roughly 200 kWh per capita in 2020, it is still around the average of the time period between 2010-2020, which means the total energy consumption even in the pandemic has stayed the same.

Analysing each sector individually, we can see that Domestic, Non-domestic, Industry and the Agricultural sectors follow the general energy consumption tendency of increasing. Unlike to what we expected, it is not apparent that there was a drastic reduction in energy consumption during the financial crisis of 2007-2008.

When observing the StreetLights and GovernmentBuildings sectors, there is a tendency of less power usage since the 2010s. In the case of StreetLights, this could be explained by the early adoption of more efficient street lighting lamps, as evidenced in some news articles in the era. In the case of GovernmentBuildings, this less usage of power is likely related to the creation of different online portals for the different governmental organizations, and the early adoption by people of those systems. A good example of this is the creation of the Portuguese Finances Portal for taxes declaration in the early 2010s.

Putting the 2020s in perspective, we can clearly observe that there was a general reduction of energy consumption in all sectors, when compared to the previous 4-5 years, with exception of the Domestic energy consumption. This likely happened, as we suspected, due to the country's Lockdown and remote work/schooling policies, in an attempt to reduce the impact of the Covid-19 pandemic.

## Predictions

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
rolling_mean = df.rolling(window = 27).mean()
rolling_std = df.rolling(window = 27).std()
```

```python
columns = ['Total', 'Domestic', 'Non-domestic', 'Industry', 'Agriculture', 'StreetLights','GovernmentBuildings']
series_alg = []

for i in columns:
    df_log = np.log(df[i])
    df_log_shift = df_log - df_log.shift()
    decomposition = seasonal_decompose(df_log, period=1) 
    model = ARIMA(df_log, order=(4, 0, 1))
    results = model.fit(disp=-1)
    #plt.plot(df_log_shift)
    #plt.plot(results.fittedvalues, color='red')
    series_alg.append(results.predict(27,32))
```

```python
series_normal = np.exp(series_alg)
years = [2020, 2021, 2022, 2023, 2024, 2025]
data = {'Year': years, 'Total': series_normal[0], 'Domestic': series_normal[1], 'Non-domestic': series_normal[2], 'Industry': series_normal[3], 'Agriculture': series_normal[4], 'StreetLights': series_normal[5], 'GovernmentBuildings': series_normal[6]}
df_new = pd.DataFrame(data)
df_new.set_index('Year', inplace=True)
df_new.drop(index=2020, inplace=True)
df_new
```

```python
df_full = pd.concat([df, df_new])
df_full
```

```python
fig, ax =plt.subplots(4,2, figsize=(20, 20))
sns.lineplot(data=df_full, x='Year', y='Total', ax=ax[0][0])
sns.lineplot(data=df_full, x='Year', y='Domestic', ax=ax[0][1])
sns.lineplot(data=df_full, x='Year', y='Non-domestic', ax=ax[1][0])
sns.lineplot(data=df_full, x='Year', y='Industry', ax=ax[1][1])
sns.lineplot(data=df_full, x='Year', y='Agriculture', ax=ax[2][0])
sns.lineplot(data=df_full, x='Year', y='StreetLights', ax=ax[2][1])
sns.lineplot(data=df_full, x='Year', y='GovernmentBuildings', ax=ax[3][0])
```

## Interpretation of forecast results and conclusion

As mentioned before, the predictions for the next 5 years are assuming that life will slowly return to a normal pre-pandemic state. In this we see a general reduction of power consumption in all areas, as some businesses have already closed, rendering to a reduction of industrial and agricultural activity, and therefore reduction of power usage.

As people slowly resume their work in the office, the power usage in homes will also likely suffer a reducion. In general, there will be a lower power consumption, specially as the government slowly switches the street lightning to more efficient technologies, as mentioned before.

We would like to reinforce the idea that this forecast is based mostly on pre-pandemic data and that life will return to a pre-pandemic life state. We would need more data for forecasting a scenario where the pandemic lasts.
